```python
# categorical
categorical_df = df.select_dtypes(include=['object', 'category'])
categorical_df.head()
```

```python
# categorical
categorical_df = df.select_dtypes(include=['object', 'category'])
categorical_df["hypertension"] = df["hypertension"]
categorical_df["stroke"] = df["stroke"]
categorical_df["heart_disease"] = df["heart_disease"]
categorical_df
```

```python
for column in categorical_df.columns:
    print(f"Value counts for {column}:")
    print(categorical_df[column].value_counts())
    print('---' * 10)
```

```python
for i in categorical_df.columns:
  counts = categorical_df[i].value_counts()

  # Plotting
  fig, axes = plt.subplots(1, 2, figsize=(12, 6))

  # Pie chart
  axes[0].pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
  axes[0].set_title(i)

  # Bar chart
  counts.plot(kind='bar', ax=axes[1])
  axes[1].set_title(i)
  axes[1].set_ylabel('Count')
  axes[1].set_xlabel(i)

  plt.tight_layout()
  plt.show()
```

```python
# categorical result
#  gender biased to the female
# gender other is only 1 value need to drop it
# ever_ married biased to the married value
#work_type biased to the private work
#work type there's 22 one not work which is a very small percentage 0.4%
#hypertension biased to 0
#stroke biased to 0
```

```python
import pandas as pd
import plotly.express as px

# Load data
df = pd.read_csv('/content/healthcare-dataset-stroke-data.csv')

# Create interactive plot
fig = px.scatter(df, x='smoking_status', y='stroke', color='smoking_status', hover_data=['age', 'hypertension','heart_disease'],
                 labels={'smoking_status': 'Smoking Status', 'stroke': 'Stroke', 'age': 'Age', 'hypertension': 'Hypertension','heart_disease':'heart_disease'},
                 title='Stroke by Smoking Status')
fig.update_traces(marker=dict(size=12))
fig.show()
```

```python
import plotly.express as px
ct = pd.crosstab(df['gender'], df['heart_disease'])

# Reset the index to make 'gender' a regular column
ct.reset_index(inplace=True)

# Melt the DataFrame to make it suitable for plotting
melted_ct = pd.melt(ct, id_vars='gender', value_vars=[0, 1],
                    var_name='heart_disease', value_name='Count')

# Create the interactive bar plot
fig = px.bar(melted_ct, x='gender', y='Count',
            color='heart_disease', title='heart disease by gender',
            labels={'gender': 'gender', 'Count': 'Number of People'},
            text='Count')

# Save the plot to an HTML file
fig.write_html("heartdisease_by_gender.html")

# Show the plot
fig.show()
```
